```python
import tkinter as tk
from tkinter import ttk, messagebox
import re
```

# function for home page 

```python
# Contact storage
contacts = []

def show_frame(frame):
    frame.tkraise()

def create_home_page(parent):
    frame = tk.Frame(parent)
    frame.grid(row=0, column=0, sticky="nsew")

    label = tk.Label(frame, text="Welcome to the Contact Management Application", font=('Helvetica', 18, 'bold'))
    label.pack(side="top", fill="x", pady=10)

    info_label = tk.Label(frame, text="Use the taskbar above to navigate through the application.\n"
                                      "You can add, view, and search for contacts easily.",
                          font=('Helvetica', 14))
    info_label.pack(side="top", fill="x", pady=10)
    
    return frame
```

# function to manage the add contact page 

```python
def create_add_contact_page(parent):
    frame = tk.Frame(parent)
    frame.grid(row=0, column=0, sticky="nsew")

    label = tk.Label(frame, text="Add New Contact", font=('Helvetica', 18, 'bold'))
    label.pack(side="top", fill="x", pady=10)

    name_var = tk.StringVar()
    phone_var = tk.StringVar()
    email_var = tk.StringVar()
    address_var = tk.StringVar()

    name_label = ttk.Label(frame, text="Name:")
    name_label.pack(pady=5)
    name_entry = ttk.Entry(frame, textvariable=name_var)
    name_entry.pack(pady=5)

    phone_label = ttk.Label(frame, text="Phone:")
    phone_label.pack(pady=5)
    phone_entry = ttk.Entry(frame, textvariable=phone_var)
    phone_entry.pack(pady=5)

    email_label = ttk.Label(frame, text="Email:")
    email_label.pack(pady=5)
    email_entry = ttk.Entry(frame, textvariable=email_var)
    email_entry.pack(pady=5)

    address_label = ttk.Label(frame, text="Address:")
    address_label.pack(pady=5)
    address_entry = ttk.Entry(frame, textvariable=address_var)
    address_entry.pack(pady=5)

    def add_contact():
        name = name_var.get()
        phone = phone_var.get()
        email = email_var.get()
        address = address_var.get()

        if not name or not phone:
            messagebox.showwarning("Input Error", "Name and Phone are required fields.")
            return
        if len(phone) != 10 or not phone.isdigit():
            messagebox.showwarning("Input Error", "Phone number must be 10 digits long.")
            return
        if not re.match(r"[^@]+@[^@]+\.[^@]+", email):
            messagebox.showwarning("Input Error", "Invalid email format.")
            return

        contact = {"name": name, "phone": phone, "email": email, "address": address}
        contacts.append(contact)
        messagebox.showinfo("Success", "Contact added successfully!")
        name_var.set("")
        phone_var.set("")
        email_var.set("")
        address_var.set("")

    add_button = ttk.Button(frame, text="Add Contact", command=add_contact)
    add_button.pack(pady=5)

    return frame
```

# function for viewing the list of contact

```python
def create_view_contact_page(parent):
    frame = tk.Frame(parent)
    frame.grid(row=0, column=0, sticky="nsew")

    label = tk.Label(frame, text="View Contacts", font=('Helvetica', 18, 'bold'))
    label.pack(side="top", fill="x", pady=10)

    tree = ttk.Treeview(frame, columns=("Name", "Phone", "Email", "Address"), show="headings")
    tree.heading("Name", text="Name")
    tree.heading("Phone", text="Phone")
    tree.heading("Email", text="Email")
    tree.heading("Address", text="Address")
    tree.pack(fill="both", expand=True)

    def refresh_list():
        for item in tree.get_children():
            tree.delete(item)

        for contact in contacts:
            tree.insert("", "end", values=(contact["name"], contact["phone"], contact["email"], contact["address"]))

    refresh_button = ttk.Button(frame, text="Refresh List", command=refresh_list)
    refresh_button.pack(pady=5)

    return frame
```

# function for search page

```python
def create_search_contact_page(parent):
    frame = tk.Frame(parent)
    frame.grid(row=0, column=0, sticky="nsew")

    # Left Frame for Searching
    search_frame = tk.Frame(frame)
    search_frame.pack(side="left", fill="both", expand=True)

    search_label = tk.Label(search_frame, text="Search Contact", font=('Helvetica', 18, 'bold'))
    search_label.pack(side="top", fill="x", pady=10)

    search_var = tk.StringVar()

    search_label = ttk.Label(search_frame, text="Search by Name or Phone:")
    search_label.pack(pady=5)
    search_entry = ttk.Entry(search_frame, textvariable=search_var)
    search_entry.pack(pady=5)

    search_button = ttk.Button(search_frame, text="Search", command=lambda: search_contact(search_var, search_tree))
    search_button.pack(pady=5)

    search_tree = ttk.Treeview(search_frame, columns=("Name", "Phone", "Email", "Address"), show="headings")
    search_tree.heading("Name", text="Name")
    search_tree.heading("Phone", text="Phone")
    search_tree.heading("Email", text="Email")
    search_tree.heading("Address", text="Address")
    search_tree.pack(fill="both", expand=True)

    # Right Frame for Update and Delete
    edit_frame = tk.Frame(frame)
    edit_frame.pack(side="right", fill="both", expand=True)

    edit_label = tk.Label(edit_frame, text="Edit Contact", font=('Helvetica', 18, 'bold'))
    edit_label.pack(side="top", fill="x", pady=10)

    name_var = tk.StringVar()
    phone_var = tk.StringVar()
    email_var = tk.StringVar()
    address_var = tk.StringVar()

    name_label = ttk.Label(edit_frame, text="Name:")
    name_label.pack(pady=5)
    name_entry = ttk.Entry(edit_frame, textvariable=name_var)
    name_entry.pack(pady=5)

    phone_label = ttk.Label(edit_frame, text="Phone:")
    phone_label.pack(pady=5)
    phone_entry = ttk.Entry(edit_frame, textvariable=phone_var)
    phone_entry.pack(pady=5)

    email_label = ttk.Label(edit_frame, text="Email:")
    email_label.pack(pady=5)
    email_entry = ttk.Entry(edit_frame, textvariable=email_var)
    email_entry.pack(pady=5)

    address_label = ttk.Label(edit_frame, text="Address:")
    address_label.pack(pady=5)
    address_entry = ttk.Entry(edit_frame, textvariable=address_var)
    address_entry.pack(pady=5)

    def search_contact(search_var, search_tree):
        query = search_var.get().lower()
        search_tree.delete(*search_tree.get_children())

        for contact in contacts:
            if query in contact["name"].lower() or query in contact["phone"]:
                search_tree.insert("", "end", values=(contact["name"], contact["phone"], contact["email"], contact["address"]))

    def on_tree_select(event):
        selected_item = search_tree.selection()
        if not selected_item:
            return

        contact_details = search_tree.item(selected_item[0], "values")
        name_var.set(contact_details[0])
        phone_var.set(contact_details[1])
        email_var.set(contact_details[2])
        address_var.set(contact_details[3])

    search_tree.bind("<<TreeviewSelect>>", on_tree_select)

    def update_contact():
        selected_item = search_tree.selection()
        if not selected_item:
            messagebox.showwarning("Selection Error", "Please select a contact to update.")
            return

        contact_index = search_tree.index(selected_item[0])
        if len(phone_var.get()) != 10 or not phone_var.get().isdigit():
            messagebox.showwarning("Input Error", "Phone number must be 10 digits long.")
            return
        if not re.match(r"[^@]+@[^@]+\.[^@]+", email_var.get()):
            messagebox.showwarning("Input Error", "Invalid email format.")
            return

        contacts[contact_index] = {
            "name": name_var.get(),
            "phone": phone_var.get(),
            "email": email_var.get(),
            "address": address_var.get()
        }
        search_contact(search_var, search_tree)
        messagebox.showinfo("Success", "Contact updated successfully!")

    def delete_contact():
        selected_item = search_tree.selection()
        if not selected_item:
            messagebox.showwarning("Selection Error", "Please select a contact to delete.")
            return

        contact_index = search_tree.index(selected_item[0])
        del contacts[contact_index]
        search_contact(search_var, search_tree)
        name_var.set("")
        phone_var.set("")
        email_var.set("")
        address_var.set("")
        messagebox.showinfo("Success", "Contact deleted successfully!")

    update_button = ttk.Button(edit_frame, text="Update Contact", command=update_contact)
    update_button.pack(pady=5)

    delete_button = ttk.Button(edit_frame, text="Delete Contact", command=delete_contact)
    delete_button.pack(pady=5)

    return frame
```

# main window of contact application 

```python
def main():
    root = tk.Tk()
    root.title("Contact Management Application")
    root.geometry("1000x500")  

    # Create the taskbar
    taskbar = tk.Frame(root, bd=1, relief=tk.RAISED)
    taskbar.pack(side="top", fill="x")

    home_button = ttk.Button(taskbar, text="Home", command=lambda: show_frame(home_frame))
    home_button.pack(side="left", padx=2, pady=2)

    add_contact_button = ttk.Button(taskbar, text="Add", command=lambda: show_frame(add_contact_frame))
    add_contact_button.pack(side="left", padx=2, pady=2)

    view_contact_button = ttk.Button(taskbar, text="View", command=lambda: show_frame(view_contact_frame))
    view_contact_button.pack(side="left", padx=2, pady=2)

    search_contact_button = ttk.Button(taskbar, text="Search", command=lambda: show_frame(search_contact_frame))
    search_contact_button.pack(side="left", padx=2, pady=2)

    # Create a container
    container = tk.Frame(root)
    container.pack(side="top", fill="both", expand=True)

    global home_frame, add_contact_frame, view_contact_frame, search_contact_frame
    home_frame = create_home_page(container)
    add_contact_frame = create_add_contact_page(container)
    view_contact_frame = create_view_contact_page(container)
    search_contact_frame = create_search_contact_page(container)

    # Show the initial page
    show_frame(home_frame)

    root.mainloop()
```

# execution of whole program 

```python
if __name__ == "__main__":
    main()
```
